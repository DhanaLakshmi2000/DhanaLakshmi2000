```
#checking for any null values
dataframe.drop(dataframe[dataframe['MSRP']==0].index,inplace=True)
```

```
#dropping the columns which are of less relevance 
dataframe.drop(['Market Category'],axis=1,inplace=True)
#as it has more NaN values i think it is less relevance
```

```
dataframe.shape
```

```
#renaming the column names
dataframe.rename(columns={"Engine HP": "HP","Engine Cylinders": "Cylinders","Transmission Type":"Transmission","Driven_wheels":"Driven mode","highway MPG":"MPG-H","city mpg":"MPG-C","MSRP":"price"},inplace = True)
dataframe
```

```
#finding the duplicates in the given file
duplicateDFROW = dataframe[dataframe.duplicated(keep='last')]
duplicateDFROW
```

```
#removing duplicates from the .csv file
dataframe.drop_duplicates(subset=None,keep='first',inplace=True)
```

```
dataframe.shape
```

```
#checking whether the duplicates are available in the file or not
duplicateDFROW = dataframe[dataframe.duplicated(keep='last')]
duplicateDFROW
```

```
#filling the NaN values for the specified indices with "ffill" 
dataframe['Engine Fuel Type'].fillna(value='ffill',inplace=True)
dataframe['HP'].fillna(value='ffill',inplace=True)
dataframe['Cylinders'].fillna(value='ffill',inplace=True)
dataframe['Number of Doors'].fillna(value='ffill',inplace=True)
```
